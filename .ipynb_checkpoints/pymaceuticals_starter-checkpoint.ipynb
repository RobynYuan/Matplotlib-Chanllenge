## Observations and Insights 

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
from scipy.stats import linregress

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"


# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)



combined_df = pd.merge(mouse_metadata,study_results, on="Mouse ID")


# Display the data table for preview
combined_df. head(20)
```

```python
# Checking the number of mice.
count=len(mouse_metadata["Mouse ID"])
count
```

```python
# Getting the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 
duplicates=combined_df[combined_df.duplicated()]
duplicates
  


      
```

```python
# Optional: Get all the data for the duplicate mouse ID. 

To_be_removed=combined_df[combined_df['Mouse ID']== "g989"]
To_be_removed
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
Clean_combined_df=combined_df[combined_df['Mouse ID'] != "g989"]
Clean_combined_df.head()
```

```python
# Checking the number of mice in the clean DataFrame.

mouses=Clean_combined_df['Mouse ID'].unique()
mouse_number=len(mouses)
mouse_number
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen


grouped_Clean_combined_df= Clean_combined_df.groupby(["Drug Regimen"])
grouped_Clean_combined_df.head()
mean=round(grouped_Clean_combined_df['Tumor Volume (mm3)'].mean(),2)
median=round(grouped_Clean_combined_df['Tumor Volume (mm3)'].median(),2)
variance=round(grouped_Clean_combined_df['Tumor Volume (mm3)'].var(),2) 
standard_deviation=round(grouped_Clean_combined_df['Tumor Volume (mm3)'].std(),2)
SEM=round(grouped_Clean_combined_df['Tumor Volume (mm3)'].sem(),2)

summary_df = pd.DataFrame({"mean": mean, "median": median,"variance":variance,"standard deviation":standard_deviation,"SEM":SEM})
summary_df



```

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin

# Start by getting the last (greatest) timepoint for each mouse

Last_timepoint=Clean_combined_df.groupby(["Mouse ID"])['Timepoint'].max()
Final_timepoint=pd.DataFrame(Clean_combined_df.groupby(["Mouse ID"])['Timepoint'].max())
Mouse_ID=Last_timepoint.index.values
final_timepoint_df=pd.DataFrame({"Mouse ID":Mouse_ID,"Timepoint":Last_timepoint})
new_final_timepoint_df=final_timepoint_df.reset_index(drop=True)
new_final_timepoint_df


# # # Merge this group df with the original dataframe to get the tumor volume at the last timepoint
final_tumor_volume = new_final_timepoint_df.merge(Clean_combined_df,on=['Mouse ID','Timepoint'],how="left")

final_tumor_volume


```

## Line and Scatter Plots

```python
# Generate a line plot of tumor volume vs. time point for a mouse treated with Capomulin

Capomulin=Clean_combined_df[Clean_combined_df['Drug Regimen']=="Capomulin"]
Capomulin_s185=Capomulin[Capomulin['Mouse ID']=="s185"]
plt.plot(Capomulin_s185["Timepoint"],Capomulin_s185["Tumor Volume (mm3)"])
plt.xlabel("Timepoint")
plt.ylabel("Tumor Volume (mm3)")
plt.title("Tumor volume vs. Time point")
plt.show()

```

```python
# Generate a scatter plot of average tumor volume vs. mouse weight for the Capomulin regimen
Capomulin.grouped=Capomulin.groupby("Mouse ID")
average=Capomulin.grouped.mean()
plt.scatter(average["Weight (g)"],average["Tumor Volume (mm3)"])
plt.xlabel("mouse weight")
plt.ylabel("average tumor volume")
plt.title("average tumor volume vs. mouse weight")
plt.show()
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and linear regression model for mouse weight and average tumor volume for the Capomulin regimen
correlation = st.pearsonr(average["Weight (g)"],average["Tumor Volume (mm3)"])
print(f"The correlation between both factors is {round(correlation[0],2)}")
```

```python
# Print out the r-squared value along with the plot.
x_values = average["Weight (g)"]
y_values = average["Tumor Volume (mm3)"]
(slope, intercept, rvalue, pvalue, stderr) = linregress(x_values, y_values)
regress_values = x_values * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
plt.scatter(x_values,y_values)
plt.plot(x_values,regress_values,"r-")
plt.annotate(line_eq,(6,10),fontsize=15,color="red")
plt.xlabel("mouse weight")
plt.ylabel("average tumor volume")
plt.title("average tumor volume vs. mouse weight")
print(f"The r-squared is: {rvalue**2}")
plt.show()
```
